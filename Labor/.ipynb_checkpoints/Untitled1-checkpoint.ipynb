```python
#Labor2
#1
#c
reset()
x = var('x')
y = function('y')(x)
_C = var('_C')
DGL = diff(y, x) == 2 * x * (1 + y ^ 2)
sol = desolve(DGL, y)
sol
```

```python
a = solve(sol, y)
a
```

```python
sol = a[0].rhs()
sol
```

```python
sol1 = sol.substitute(_C == 1)
g = plot(sol1, x, -0.5, 0.5)

for i in [11..20]:
    sol1 = sol.substitute(_C == i / 10)
    g += plot(sol1, x, -0.5, 0.5)
show(g)
```

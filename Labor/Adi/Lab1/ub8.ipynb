```sagemath
x,y = var('x,y')
f(x) = x^2 + y^2 - 2*x - 4*y + 4
implicit_plot(f(x,y) == 0, (x, -1, 3), (y, -1, 3))
```

```sagemath
x,y = var('x,y')
f(x, y) = x^2 - 2*x*y - y^2
implicit_plot(f(x,y) == 1, (x, -1, 1), (y, -1, 1))
```

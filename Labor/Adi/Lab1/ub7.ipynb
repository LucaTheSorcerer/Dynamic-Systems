```sagemath
t = var('t')
x(t) = (1 - cos(t)) * cos(t)
y(t) = (1 - cos(t)) * cos(t)
parametric_plot((x(t), y(t)), (t, 0, 2*pi))
```

```sagemath
t = var('t')
x(t) = sin(3*t)*cos(t)
y(t) = sin(3*t)*sin(t)
parametric_plot((x(t), y(t)), (t, 0, 2*pi))
```

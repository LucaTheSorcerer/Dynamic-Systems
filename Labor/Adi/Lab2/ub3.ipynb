```sagemath
#a)
x,k = var('x,k')
y = function('y')(x)
eq = diff(y,x) + (k/x)*y == x^3
desolve(eq,[y,x])
```

```sagemath
#b)
x = var('x')
y = function('y')(x)
eq = diff(y,x) + (1/x)*y == x^3
sol = desolve(eq,y)
_C=var('_C')
sol1 = sol.substitute(_C==1)
plot(sol1,x,-1,1)
```

```sagemath
#c)
x = var('x')
y = function('y')(x)
eq = diff(y,x) + (1/x)*y == x^3
sol = desolve(eq,y,ics=[1,0])
_C = var('_C')
sol1 = sol.substitute(_C == 1)
plot(sol1,x,-1,1)
```

```sagemath
#d)
x,k = var('x,k')
y = function('y')(x)
eq = diff(y,x) + (k/x)*y == x^3
desolve(eq,[y,x])
sol(x,k) = desolve(eq,[y,x])
sol(x,k)
```

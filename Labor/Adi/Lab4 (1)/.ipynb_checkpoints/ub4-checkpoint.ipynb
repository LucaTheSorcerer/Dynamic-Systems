```python
#a)
f1(u,v)=2*u+v
f2(u,v)=u+2*v
n=sqrt(f1(u,v)^2+f2(u,v)^2)
plot_vector_field([f1(u,v)/n,f2(u,v)/n],[u,-10,10],[v,-10,10])
```

```python
#b)
f1(u,v) = v
f2(u,v) = -u
n=sqrt(f1(u,v)^2 + f2(u,v)^2)
plot_vector_field([f1(u,v)/n,f2(u,v)/n],[u,-10,10],[v,-10,10])
```

```python
#c)
f1(u,v) = -u-v
f2(u,v) = u-v
n = sqrt(f1(u,v)^2 + f2(u,v)^2)
plot_vector_field([f1(u,v)/n,f2(u,v)/n],[u,-10,10],[v,-10,10])
```

```python
#d)
f1(u,v) = -2*u
f2(u,v) = -4*u - 2*v
n = sqrt(f1(u,v)^2 + f2(u,v)^2)
plot_vector_field([f1(u,v)/n,f2(u,v)/n],[u,-10,10],[v,-10,10])
```

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x+4*y
deq2=diff(y,t)==x+y
syst=[deq1,deq2]
desolve_system(syst,[x,y])
```

```python
C1,C2=var('C1,C2')
desolve_system(syst, [x,y],[0,C1,C2])
```

```python
sol=desolve_system(syst, [x,y],[0,C1,C2])
sol_x(t,C1,C2)=sol[0].rhs()
sol_x
```

```python
sol_y(t,C1,C2)=sol[1].rhs()
sol_y
```

```python
plot(sol_x(t,1,1),t,-4,4,color='red')
```

```python
plot(sol_y(t,1,1),t,-4,4,color='blue')
```

```python
g1=plot(sol_x(t,1,1),t,-4,4,color='red')
g2=plot(sol_y(t,1,1),t,-4,4,color='blue')
show(g1+g2)
```

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
z=function('z')(t)
deq1=diff(x,t)==x-y+z
deq2=diff(y,t)==x+y-z
deq3=diff(z,t)==-y+2*z
syst=[deq1,deq2,deq3]
desolve_system(syst,[x,y,z])
```

```python
C1,C2,C3=var('C1,C2,C3')
desolve_system(syst, [x,y,z],[0,C1,C2,C3])
```

```python
sol=desolve_system(syst, [x,y,z],[0,C1,C2,C3])
sol_x(t,C1,C2,C3)=sol[0].rhs()
sol_x
```

```python
sol_y(t,C1,C2,C3)=sol[1].rhs()
sol_y
```

```python
sol_z(t,C1,C2,C3)=sol[2].rhs()
sol_z
```

```python
plot(sol_x(t,1,1,1),t,-4,4,color='red')
```

```python
plot(sol_y(t,0,2,3),t,-4,4,color='blue')
```

```python
plot(sol_z(t,-2,1,5),t,-4,4,color='green')
```

```python
g1=plot(sol_x(t,1,1,1),t,-4,4,color='red')
g2=plot(sol_y(t,0,2,3),t,-4,4,color='blue')
g3=plot(sol_z(t,-2,1,5),t,-4,4,color='green')
show(g1+g2+g3)
```

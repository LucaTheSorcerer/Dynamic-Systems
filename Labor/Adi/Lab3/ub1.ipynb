```python
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)+y==sin(x)+cos(x)
sol=desolve(eqd,y)
sol
```

```python
_K1,_K2=var('_K1,_K2')
sol.substitute(_K1==1,_K2==3)
```

```python
sol1=sol.substitute(_K1==1,_K2==3)
plot(sol1,x,-2,2)
```

```python
eqd=diff(y,x,2)-y==e^2*x
desolve(eqd,y)
```

```python
_K1,_K2=var('_K1,_K2')
sol.substitute(_K1==1,_K2==3)
```

```python
sol2=sol.substitute(_K1==0,_K2==1)
plot(sol2,x,-1.5,3)
```

```python
eqd=diff(y,x,2)+4*y==1/cos(2*x)
desolve(eqd,y)
```

```python
sol3=sol.substitute(_K1==2, _K2==3)
plot(sol3,x,2,5)
```

```python
eqd=diff(y,x,2)-diff(y,x)==1/(1+e^x)
desolve(eqd,y)
```

```python
sol4=sol.substitute(_K1==1,_K2==4)
plot(sol4,x,1,3)
```

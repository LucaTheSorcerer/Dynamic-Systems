```python
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)-5*diff(y,x)+4*y==0
sol1=desolve(eqd,y,ics=[0,5,8])
sol1
```

```python
plot(sol1,x,-2,2)
```

```python
eqd=diff(y,x,2)-4*diff(y,x)+5*y==2*(x^2)*(e^x)
sol2=desolve(eqd,y,ics=[0,2,3])
sol2
```

```python
plot(sol2,x,-2,2)
```

```python
eqd=diff(y,x,2)+4*diff(y,x)==4*(sin(2*x)+cos(2*x))
sol3=desolve(eqd,y,ics=[pi,2*pi,2*pi])
sol3
```

```python
plot(sol3,x,-2,2)
```

```python
#a) wir mussen zuerst k mit 0.5 und x0 mit 10 einsetzen und dann das System losen
t=var('t')
x=function('x')(t)
k=0.5
deq=diff(x,t)==-k*x
sol=desolve(deq,x,ics=[0,10])
plot(sol,(t,0,10))
```

```python
#a)
t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==3*x+y
deq2=diff(y,t)==x+3*y
syst=[deq1,deq2]
desolve_system(syst,[x,y])
```

```python
C1,C2=var('C1,C2')
desolve_system(syst, [x,y],[0,C1,C2])
```

```python
sol=desolve_system(syst, [x,y],[0,C1,C2])
sol_x(t,C1,C2)=sol[0].rhs()
sol_x
```

```python
sol_y(t,C1,C2)=sol[1].rhs()
sol_y
```

```python
#b)
t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x+4*y
deq2=diff(y,t)==x+y
syst=[deq1,deq2]
vars=[x,y]
in_cond=[0,3,3]
```

```python
sol=desolve_system(syst, vars,in_cond)
sol
```

```python
sol_x(t)=sol[0].rhs()
sol_x
```

```python
sol_y(t)=sol[1].rhs()
sol_y
```

```python
#a)
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)+2*diff(y,x)-3*y==0
sol=desolve(eqd,y)
sol
```

```python
_K1,_K2=var('_K1,_K2')
sol.substitute(_K1==1,_K2==3)
```

```python
sol1=sol.substitute(_K1==1,_K2==3)
plot(sol1,x,-2,2)
```

```python
#b)
x,a=var('x,a')
y=function('y')(x)
eqd=diff(y,x,2)+2*diff(y,x)-3*y==0
sol=desolve(eqd,y,ics=[0,a,1])
sol
```

```python
sol=sol.substitute(a=1)
limit(sol,x=infinity)
```

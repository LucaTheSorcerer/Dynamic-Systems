```python
#a)
x=var('x')
y=function('y')(x)
eqd=diff(y,x)+y*tan(x)==1/cos(x)
sol=desolve(eqd,y)
sol
```

```python
_C=var('_C')
sol.substitute(_C==1)
```

```python
sol1=sol.substitute(_C==1)
plot(sol1,x,-2,2)
```

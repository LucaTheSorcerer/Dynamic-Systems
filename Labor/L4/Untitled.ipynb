```python
#a.
# x' = x + 4y
# y' = x + y
# x(0) = 1
# y(0) = 2

t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x + 4 * y
deq2=diff(y,t)==x + y
syst=[deq1,deq2]
vars=[x,y]
in_cond=[0,1,2]
```
